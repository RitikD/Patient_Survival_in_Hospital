```python
import pandas as pd  
import numpy as np

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
pharma_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')
```

```python
pharma_data.head()
```

```python
pharma_data['Treated_with_drugs'] = pharma_data['Treated_with_drugs'].fillna(pharma_data['Treated_with_drugs'].mode()[0])
pharma_data.head()
```

```python
df = pd.DataFrame(pharma_data)
df['Treated_with_drugs'].str.get_dummies(sep=' ')
```

```python
pharma_data = pd.get_dummies(pharma_data, columns=['Treated_with_drugs'])
```

```python
pharma_data['Patient_mental_condition'] = pharma_data['Patient_mental_condition'].fillna(pharma_data['Patient_mental_condition'].mode()[0])
pharma_data.head()
```

```python
df = pd.DataFrame(pharma_data)
df['Patient_mental_condition'].str.get_dummies(sep=' ')
```

```python
pharma_data = pd.get_dummies(pharma_data, columns=['Patient_mental_condition'])
```

```python
pharma_data['Patient_Smoker'] = pharma_data['Patient_Smoker'].fillna(pharma_data['Patient_Smoker'].mode()[0])
pharma_data['Patient_Smoker'].replace(['Cannot say'],'NO', inplace = True)
pharma_data.head()
```

```python
df = pd.DataFrame(pharma_data)
df['Patient_Smoker'].str.get_dummies(sep=' ')
```

```python
pharma_data = pd.get_dummies(pharma_data, columns=['Patient_Smoker'])
```

```python
pharma_data['Patient_Rural_Urban'] = pharma_data['Patient_Rural_Urban'].fillna(pharma_data['Patient_Rural_Urban'].mode()[0])
pharma_data.head()
```

```python
df = pd.DataFrame(pharma_data)
df['Patient_Rural_Urban'].str.get_dummies(sep=' ')
```

```python
pharma_data = pd.get_dummies(pharma_data, columns=['Patient_Rural_Urban'])
```

```python
#pharma_data.isnull().sum()
```

```python
#x = pharma_data.drop('Survived_1_year', axis = 1)
```

```python
pharma_data['A'] = pharma_data['A'].fillna(pharma_data['A'].mode()[0])
pharma_data['B'] = pharma_data['B'].fillna(pharma_data['B'].mode()[0])
pharma_data['C'] = pharma_data['C'].fillna(pharma_data['C'].mode()[0])
pharma_data['D'] = pharma_data['D'].fillna(pharma_data['D'].mode()[0])
pharma_data['E'] = pharma_data['E'].fillna(pharma_data['E'].mode()[0])
pharma_data['F'] = pharma_data['F'].fillna(pharma_data['F'].mode()[0])
pharma_data['Z'] = pharma_data['Z'].fillna(pharma_data['Z'].mode()[0])
pharma_data['Number_of_prev_cond'] = pharma_data['A']+pharma_data['B']+pharma_data['C']+pharma_data['D']+pharma_data['E']+pharma_data['F']+pharma_data['Z']
```

```python
pharma_data.head()
```

```python
#Gio moves this
X = pharma_data.drop('Survived_1_year', axis = 1)
y = pharma_data['Survived_1_year']
```

```python
#Gio moves this
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
logisticRegr = LogisticRegression(solver= 'liblinear')
logisticRegr.fit(X_train, y_train)
```

```python
pred_train = logisticRegr.predict(X_train)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_train, pred_train)
cnf_matrix
```

```python
pred = logisticRegr.predict(X_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, pred)
cnf_matrix
```
